```python
import numpy as np
import soydata
from soydata.data.clustering import make_swiss_roll
from soydata.visualize import use_notebook
from soydata.visualize import scatterplot3d

use_notebook()
print(f'soydata=={soydata.__version__}')

X, colors = make_swiss_roll(n_samples=3000, n_rotations=3, 
    gap=0.5, thickness=0.0, width=10.0, discretize=True)
```

```python
colors_ = colors.copy()
indices = np.where(colors >= 128)[0]
colors_[indices] = 256 - colors[indices]

fig2 = scatterplot3d(X, colors_, marker_size=2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression

regressor_names = ['knn', 'mlp', 'boosting', 'linear']
regressors = [
    KNeighborsRegressor(),
    MLPRegressor(hidden_layer_sizes=(5, 2), max_iter=1000),
    GradientBoostingRegressor(),
    LinearRegression()
]

color_preds = []
for regressor in regressors:
    regressor.fit(X, colors_)
    color_pred = regressor.predict(X)
    color_preds.append(color_pred)
```

```python
for name, color_pred in zip(regressor_names, color_preds):
    fig = scatterplot3d(X, color_pred, marker_size=2, title=f'Predicted with {name}')
```
